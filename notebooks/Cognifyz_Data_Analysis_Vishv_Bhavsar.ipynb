```python
# ==========================================
# COGNIFYZ DATA ANALYSIS INTERNSHIP PROJECT
# Restaurant Dataset - Full Analysis
# ==========================================

# 1️⃣ Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)

# 2️⃣ Load Dataset
df = pd.read_csv("data/Dataset.csv")

print("Dataset Loaded Successfully ✅")
print("\nFirst 5 Rows:")
display(df.head())

# 3️⃣ Basic Information
print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nStatistical Summary:")
display(df.describe())

# 4️⃣ Data Cleaning
# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Drop rows where rating is 0
df = df[df['Aggregate rating'] > 0]

print("\nAfter Cleaning:")
print("Shape:", df.shape)

# ==========================================
# 📊 EXPLORATORY DATA ANALYSIS (EDA)
# ==========================================

# 5️⃣ Top 10 Restaurants by Rating
top_restaurants = df.sort_values(by="Aggregate rating", ascending=False).head(10)

print("\nTop 10 Restaurants by Rating:")
display(top_restaurants[['Restaurant Name', 'City', 'Aggregate rating', 'Votes']])

# 6️⃣ Most Common Cuisines
cuisine_counts = df['Cuisines'].value_counts().head(10)

plt.figure()
sns.barplot(x=cuisine_counts.values, y=cuisine_counts.index)
plt.title("Top 10 Most Common Cuisines")
plt.xlabel("Count")
plt.ylabel("Cuisine")
plt.show()

# 7️⃣ City-wise Restaurant Count
city_counts = df['City'].value_counts().head(10)

plt.figure()
sns.barplot(x=city_counts.values, y=city_counts.index)
plt.title("Top 10 Cities by Restaurant Count")
plt.xlabel("Number of Restaurants")
plt.ylabel("City")
plt.show()

# 8️⃣ Price Range Distribution
plt.figure()
sns.countplot(x="Price range", data=df)
plt.title("Price Range Distribution")
plt.show()

# 9️⃣ Online Delivery Analysis
online_delivery = df['Has Online delivery'].value_counts()

plt.figure()
sns.barplot(x=online_delivery.index, y=online_delivery.values)
plt.title("Online Delivery Availability")
plt.xlabel("Online Delivery")
plt.ylabel("Count")
plt.show()

# 🔟 Rating Distribution
plt.figure()
sns.histplot(df['Aggregate rating'], bins=20, kde=True)
plt.title("Rating Distribution")
plt.xlabel("Aggregate Rating")
plt.show()

# 1️⃣1️⃣ Votes vs Rating Correlation
plt.figure()
sns.scatterplot(x="Votes", y="Aggregate rating", data=df)
plt.title("Votes vs Rating")
plt.show()

correlation = df[['Votes','Aggregate rating']].corr()
print("\nCorrelation Between Votes & Rating:")
print(correlation)

# 1️⃣2️⃣ Country-wise Restaurant Count
country_counts = df['Country Code'].value_counts().head(10)

plt.figure()
sns.barplot(x=country_counts.index, y=country_counts.values)
plt.title("Country-wise Restaurant Count")
plt.xlabel("Country Code")
plt.ylabel("Number of Restaurants")
plt.show()

# ==========================================
# 📌 Key Insights
# ==========================================

print("\n🔎 KEY INSIGHTS:")
print("1. Most restaurants fall in mid price range.")
print("2. Higher votes generally correlate with higher ratings.")
print("3. Few cuisines dominate the dataset.")
print("4. Online delivery is not available for many restaurants.")
print("5. Some cities have significantly higher restaurant density.")

print("\n✅ Analysis Completed Successfully!")
```
